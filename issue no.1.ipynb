## i/x= 4^x ?

```python
i=-0.5

while i < 1.0:
    
    a=1 / i
    b=pow(4,i)

    if a==b:
        print('i:',i,'\n')
        print('a:',a,'\n','b:',b,'\n')
        break
    elif i==0:
        i += 0.1
    else:
        i += 0.1
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return pow(4,x)


x_values = np.linspace(-1, 1, 100)


y_values = f(x_values)


plt.plot(x_values, y_values, label='y = 4^x')
plt.title('Graph of y = 4^x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        return 1/x
    

x_values = np.linspace(-1, 1, 101) 


y_values = f(x_values)


plt.plot(x_values, y_values, label='y = 1/x')
plt.title('Graph of y = 1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        return 1/x

def g(x):
    return pow(4,x)



x_values = np.linspace(-1, 1, 101)


y_values = f(x_values)
z_values = g(x_values)


plt.plot(x_values, y_values, label='y1 = 1/x')
plt.plot(x_values, z_values, label='y2 = 4^x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
