```python
# Import libraries
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Path of the file to read
museum_visitors_filepath = "/Data Science/Projects/5. Museum_Visitors/museum_visitors.csv"

# Read the file into a variable flight_data
museum_visitors_data = pd.read_csv(museum_visitors_filepath, index_col="Date")

# Prepration data
museum_visitors_pre_2016=museum_visitors_data.iloc[24:36]
museum_visitors_pre_2017=museum_visitors_data.iloc[36:48]
```

```python
museum_visitors_pre_2016
```

```python
# Print the data
museum_visitors_pre_2017
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("2016")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=museum_visitors_pre_2016.index, y=museum_visitors_pre_2016['Avila Adobe'])

# Add label for vertical axis
plt.ylabel("Number Of Visitors")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("2017")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=museum_visitors_pre_2017.index, y=museum_visitors_pre_2017['Avila Adobe'])

# Add label for vertical axis
plt.ylabel("Number Of Visitors")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("2016")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=museum_visitors_pre_2016, annot=True, fmt=".2f")

# Add label for horizontal axis
plt.xlabel("Number Of Visitors")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("2017")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=museum_visitors_pre_2017, annot=True, fmt=".2f")

# Add label for horizontal axis
plt.xlabel("Number Of Visitors")
```
